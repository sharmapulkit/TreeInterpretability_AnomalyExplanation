```python
X = data[feature_columns]
distortions = []
K = range(45,45)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

## Split into train and test templates

```python
def split_trainAndtest(ipfile, outTrainDir, outTestDir, trainFrac=0.6):
    df = pd.read_csv(ipfile)
    df = shuffle(df, random_state = shuffleSeed)
    df_train = df[:int(trainFrac*len(df))]
    df_test = df[int(trainFrac*len(df)):]
    
    df_train.to_csv(outTrainDir + os.path.basename(ipfile)[:-4] + '_train.csv')
    df_test.to_csv(outTestDir + os.path.basename(ipfile)[:-4] + '_test.csv')
```

```python
shap_max_idxs = []
ti_max_idxs = []

variance_shap = []
variance_ti = []

variance_shap_top3 = []
variance_ti_top3 = []

for _shap_file in shap_files:
    _comb = _shap_file.strip('interpreted_SHAP_outs_mb').strip('_test.txt')
    _feature_file = test_datapath + 'covComb' + _comb + '_test.csv'
    
    _ti_file = interpretation_path + 'interpreted_TI_outs_mb' + _comb + '_test.txt'
    _shap_values = np.loadtxt(interpretation_path + _shap_file)
    _ti_values = np.loadtxt(_ti_file)
    
    _shap_max_idxs = np.zeros(len(feature_columns))
    _ti_max_idxs = np.zeros(len(feature_columns))

#     print(len(_shap_values), _shap_values)
    if (len(_shap_values.shape) > 1):
        for row_id, row in enumerate(_shap_values):
            _shap_max_idxs[np.argmax(row)] += 1
            _ti_max_idxs[np.argmax(_ti_values[row_id])] += 1

            variance_shap.append(np.var(row))
            variance_ti.append(np.var(_ti_values[row_id]))

            variance_shap_top3.append(np.var(np.sort(row[:-3])))
            variance_ti_top3.append(np.var(np.sort(_ti_values[row_id][:-3])))



        shap_max_idxs.append(_shap_max_idxs)
        ti_max_idxs.append(_ti_max_idxs)
```

## Count the number of disagreements between SHAP and TI

```python
num_disagreements = 0
level_disagreements = 0
dispersion_scores_shap = []
dispersion_scores_ti = []
for template_id, (template_shap, template_ti) in enumerate(zip(shap_max_idxs, ti_max_idxs)):
    if not (np.argmax(template_shap) == np.argmax(template_ti)):
        num_disagreements += 1
#     print(template_shap)
    level_disagreements += max(template_shap) - max(template_ti)
    
    ####### Compute formant dispersion_score #######
    sorted_template_shap = template_shap.copy()
    sorted_template_shap.sort()
    max_template_shap = sorted_template_shap[-1]
    second_max_template_shap = sorted_template_shap[-2]
#     dispersion_scores_shap.append(max_template_shap/(second_max_template_shap + 1))
    dispersion_scores_shap.append((second_max_template_shap)/max_template_shap)
    
    sorted_template_ti = template_ti.copy()
    sorted_template_ti.sort()
    max_template_ti = sorted_template_ti[-1]
    second_max_template_ti = sorted_template_ti[-2]
    dispersion_scores_ti.append((second_max_template_ti)/max_template_ti)
    
dispersion_scores_shap = np.array(dispersion_scores_shap)
dispersion_scores_ti = np.array(dispersion_scores_ti)
# print(level_disagreements)
print(num_disagreements)
print("SHAP:", np.sum(dispersion_scores_shap)/len(dispersion_scores_shap))
print("TI:", np.sum(dispersion_scores_ti)/len(dispersion_scores_ti))
```

## Compute Accuracy

```python
def print_set_overlap(ranking1, ranking2, k):
    """
    Provide the correlation analysis of top-k features from the 2 list of rankings
    ranking1, ranking2: list of size = number of data points
            each element is a ranking of features
            
    returns pd.Series(rank based overlap)
    Prints the Series analysis of rank correlation values between both the rankings
    (Mean, std, Min, 25%, 50%, 75%, Max, Histogram)
    """
    overlap_values = []
    idx_arr = np.arange(len(ranking1))
    for r1, r2 in zip(ranking1, ranking2):
        
        # Sorting the rankings in descending order of the absolute attribution values.
        contributions = set([attr_value[1] for attr_value in 
                        sorted(zip(r1, idx_arr), key=lambda temp: -abs(temp[0]))][:k])
    
        overlap_len = len(contributions.intersection(r2))
        if(len(r2)==0):
            overlap = 1
        else:
            overlap = overlap_len/len(r2)
        
        overlap_values.append(overlap)
#     print(pd.Series(overlap_values).describe())
    plt.hist(overlap_values) 
    return overlap_values



def calculate_accuracy(SHAPminusBaseline, TIminusBaseline, data_x, Baseline_x):
    treatment_vars = [10,11,12]    
    treatment_vars_data = [4,5,6]
    SHAPminusBaseline_df = pd.read_csv(SHAPminusBaseline,header=None,delim_whitespace=True)
    TIminusBaseline_df = pd.read_csv(TIminusBaseline,header=None,delim_whitespace=True)
    data_x_df = pd.read_csv(data_x)
    Baseline_x_arr = np.array(pd.read_csv(Baseline_x,header=None)[0].tolist())
    
    # Getting just the treatment values from the x variables
    data_x_df_treatment = data_x_df.iloc[:,treatment_vars_data]
    data_x_df_treatment.columns = treatment_vars
    
    # Contains false if the treatment variable is not the same as the baseline
    treatment_df = (Baseline_x_arr[treatment_vars].astype(int) == data_x_df_treatment) 
    
    aList = []
    for _,row in treatment_df.iterrows():
        
        treated = set()
        for col in treatment_df:
            if row[col]==False:
                treated.add(col)
        aList.append(treated)
    return print_set_overlap(TIminusBaseline_df.to_numpy(),aList,3)

def calculate_accuracy2(SHAPminusBaseline, TIminusBasline, testData_x, Baseline_x):
    """
    SHAPminusBaseline_filename: Contains SHAP attributions (test - Baseline)
    TIminusBaseline_filename: Contains TI attributions (test - Baseline)
    test_x: Features for every test data point
    Baseline_x: Features for Baseline data point
    
    ### Higher if Intersection of top 3 features in test_x is higher with treatment variables that were flipped
    """
    SHAPminusBaseline_df = pd.read_csv(SHAPminusBaseline,header=None,delim_whitespace=True)
    TIminusBaseline_df = pd.read_csv(TIminusBaseline,header=None,delim_whitespace=True)
    testData_x_df = pd.read_csv(testData_x)
    Baseline_x_arr = np.array(pd.read_csv(Baseline_x,header=None)[0].tolist())
    
    #### Get the feature variable values from each testData point
    testData_feats = testData_x_df.loc[:, feature_columns].values
    
    #### Contains false if the treatment Variable is not the same as the baseline
    featDiffMask = np.where(testData_feats != Baseline_x)
    
    overlap_values = []
    idx_arr = np.arange(len(ranking1))
    for dataPointId, dataPoint in enumerate(featDiffMask):
        for r1, r2 in zip(dataPoint.to_numpy(), featDiffMask):
            # Sorting the rankings in descending order of the absolute attribution values.
            contributions = set([attr_value[1] for attr_value in 
                                sorted(zip(r1, idx_arr), key=lambda temp: -abs(temp[0]))][:k])

            overlap_len = len(contributions.intersection(r2))
            if(len(r2)==0):
                overlap = 1
            else:
                overlap = overlap_len/len(r2)
        
# not(1)
    
#     aList = []
#     for _, row in treatment_Mask.iterrows():
#         treated = set()
#         for col in treatment_Mask:
#             if (row[col] == False):
#                 treated.add(col)
#         aList.append(treated)
    
#     return print_set_overlap(TIminusBaseline_df.to_numpy(), aList, 3)
```

```python
runtimePath = "../../../garage/TimingAnalysis/Depth/Runtime/"
runtimeDepthsTI = {}
runtimeDepthsSHAP = {}
for depth in range(5, 60, 5):
    tiruntimeDepthNumDataPoints = []
    shapruntimeDepthNumDataPoints = []
    for datapoints in [50, 100, 200, 300, 500, 700, 900]:
        filename = runtimePath + 'runtime_MaxDepth'+str(depth)+'_NumDataPoints'+str(datapoints)
        if os.path.exists(filename):
            with open(filename, 'r') as f:
                lines = f.readlines()
                tiTime = float(lines[0].rstrip().split(':')[1])
                shapTime = float(lines[1].rstrip().split(':')[1])

                tiruntimeDepthNumDataPoints.append(tiTime)
                shapruntimeDepthNumDataPoints.append(shapTime)
    runtimeDepthsTI[depth] = tiruntimeDepthNumDataPoints
    runtimeDepthsSHAP[depth] = shapruntimeDepthNumDataPoints
    
print(runtimeDepthsSHAP[20])
```
